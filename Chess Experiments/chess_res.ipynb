```python
import pandas as pd
```

```python
df = pd.read_csv('H1H1D3.csv').drop(columns=['Unnamed: 0'])
df
```

```python
# average time to conclude a game
game_times = pd.DataFrame(columns=['TIME PER GAME'])

for i in range(0, len(df.index)-1, 2):
    game_times.loc[i, 'TIME PER GAME'] = df.iloc[i, 1] + df.iloc[i+1, 1]
    
game_times
```

```python
game_times['TIME PER GAME'].mean()
```

```python
game_times['TIME PER GAME'].std()
```

```python
# average number of moves/game
game_moves = pd.DataFrame(columns=['MOVES PER GAME'])

for i in range(0, len(df.index)-1, 2):
    game_moves.loc[i, 'MOVES PER GAME'] = df.loc[i, '# MOVES'] + df.loc[i+1, '# MOVES']
    
game_moves
```

```python
game_moves['MOVES PER GAME'].mean()
```

```python
game_moves['MOVES PER GAME'].std()
```

```python
# effective branching factor
from scipy.optimize import root_scalar

# Define the function to find the root of
def equation(x, N):
    return 1 + x + x**2 + x**3 - N - 1

# Number of nodes explored and depth of search
D = 3     # Depth of search

# Solve the equation to find the branching factor x
N = pd.DataFrame(df[df['PLAYER'] == 0]['# EXPLORED']/df[df['PLAYER'] == 0]['# MOVES'], columns=['STATE-MOVE EST. (s)']).describe().iloc[1:].iloc[0].iloc[0]

result1 = root_scalar(equation, args=(N,), bracket=[0, N])  # Provide a bracket for the solution

N = pd.DataFrame(df[df['PLAYER'] == 1]['# EXPLORED']/df[df['PLAYER'] == 1]['# MOVES'], columns=['STATE-MOVE EST. (s)']).describe().iloc[1:].iloc[0].iloc[0]
result2 = root_scalar(equation, args=(N,), bracket=[0, N])

# Check if a solution was found
if result1.converged:
    x1 = result1.root
    print(f"Estimated branching factor (x) for player 0: {x1:.2f}")
else:
    print("No solution found within the specified bracket.")


if result2.converged:
    x2 = result2.root
    print(f"Estimated branching factor (x) for player 1: {x2:.2f}")
else:
    print("No solution found within the specified bracket.")
```
